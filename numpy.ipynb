### 2. Criando Arrays(Vetores e Matrizes)

```python
# Criando um array 1D (Vetor) a partir de uma lista Pyton
vetor = np.array([1, 2, 3, 4, 5])
print("Vetor 1D:", vetor)

# Criando um array 2D (Matriz) a partir de uma lista de listas
matriz = np.array([[1, 2, 3],[4, 5, 6]])
print("\nMatriz 2D:\n", matriz)

```

O numpy possui funções nativas para gerar arrays comuns rapidamente, o que é muito útil para inicializar pesos em redes neurais, por exemplo.

```python
# Array de zeros
zeros = np.zeros((2, 3)) # 2 linhas, 3 colunas
print("Zeros:\n", zeros)
```

```python
#Array de uns
uns = np.ones((3, 3))
print("Uns:\n", uns) 
```

```python
#Array com uma sequência de números (similar ao range() do Python)
sequencia = np.arange(0, 10, 2) # Inicio, Fim(exclusivo), Passo
print("\nSequência(arange):", sequencia)
```

```python
# Valores Linearmente espaçados
espacados = np.linspace(0, 1, 5) # 5 valores entre 0 e 1
print("\nLinspace:", espacados)
```
