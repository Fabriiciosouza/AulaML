```python
import numpy as np

#1. Lendo os dados do arquivo com numpy
#pulando o cabeçalho (skip_headers)
#Extraímos APENAS AS COLUNAS DE INTERESSE para evitar erro com strings:
#2. price (y), 3: bedrooms (X1), 4: bathrooms (X2), 5: sqft_living (X3)
data = np.genfromtxt('kc_house_data.csv', delimiter=',', skip_header=1, usecols=(2, 3, 4, 5, 15, 16))

#2. Separando a variavel alvo (y) e as variaveis preditoras (X)
y = data[:, 0]  # O preço é a primeira coluna do nosso recorte
X_feature = data[:, 1:]  # O resto são as features (bedrooms, bathrooms, sqft_living)

#3. Adicionando o termo de intercepto (Viés/Bias)
# Precisamos adicionar uma coluna de numeros "1" no inicio da matriz x
# Isso representa o beta_0 na equação (o valor base da casa se tudo fosse zero)
m = X_feature.shape[0]  # Número de linhas (exemplos)
X = np.c_[np.ones(m), X_feature]  # Adiciona uma coluna de 1s no início da matriz x

#4. Aplicando a equação normal: Beta = (X.T * X)^(-1) * X.T * y
X_T = X.T # Transposta de X
# np.linalg.inv() calcula a inversa de uma matriz
# o metodo .dot() ou o operador @ realiza a multiplicação de matrizes
beta = np.linalg.inv(X_T @ X) @ X_T @ y

#5. Imprimindo os coeficientes
print("=== Coeficientes do Modelo ===")
print(f"Intercepto (Beta_0): {beta[0]:.2f}")
print(f"Coeficiente para Quartos (Beta_1): {beta[1]:.2f}")
print(f"Coeficiente para Banheiros (Beta_2): {beta[2]:.2f}")
print(f"Coeficiente para Área/sqft (Beta_3): {beta[3]:.2f}")
print(f"Coeficiente para yr_built (Beta_4): {beta[4]:.2f}")
print(f"Coeficiente para yr_renovated (Beta_5): {beta[5]:.2f}")

# 5. Fazendo uma previsão
#Vamos prever o valor da primeira casa do dataset: 3 quartos, 1 banheiro, 1180 sqft
casa_teste = np.array([1, 3, 1, 1180, 1915, 0])  # Lembre-s do "1" inicial para o intercepto!
preco_previsto = casa_teste @ beta

print("===Teste de Previsão===")
print(f"Preço REAL da casa 1: ${y[0]:.2f}")
print(f"Preço PREVISTO pelo nosso modelo: ${preco_previsto:.2f}")
```
