```python
import pandas as pd
import numpy as np #frequentimente usamos junto com o Pandas

print(f"Versão do Pandas: {pd.__version__}")
```

Series: É uma matriz unidimensional com rótulos(indices), Pode ser vista como uma unica coluna de uma tabela.

DataFrame: É uma estrutura bidimensional(Linhas e colunas), como uma pilha completa. É essencialmebte um conjunto de series.

```python
# Criando uma Series
notas = pd.Series([8.5, 9.0, 7.5, 10.0], index=['Aluno A', 'Aluno B', 'Aluno C', 'Aluno D'])
print("Exemplo de Series:\n", notas, "\n")
```

```python
# Criando um DataFrame a partir de um dicionário(útil para testes locais)
dados_turma = {
    'Nome': ['Ana', 'Bruno', 'Carlos', 'Diana'],
    'Idade': [20, 22, 21, 23],
    'Nota_Final': [8.5, 7.0, 9.5, 6.5],
    'Aprovado': [True, True, True, False]

    df = pd.DataFrame(dados_turma)
    print("Exemplo de DataFrame")
}
```

```python
# O comando abaixo é o padrão para ler arquivos CSV
#df_csv = pd.read_csv('caminho/para/seuarquivo.csv')

#Exemplo pratico carregando um datasset público diretamnete de uma URL para demonstração:
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df_titanic = pd.read_csv(url)
print("Dados carregados com sucesso!")
```
